pie plot

```python
import matplotlib.pyplot as plt
x=[40,50,60,70,80]
y=["C","C++","Java","Python","Javascript"]
plt.pie(x,labels=y)
plt.show()
```

```python
x=[40,50,60,70,80]
y=["C","C++","Java","Python","Javascript"]
ex=[0.0,0.0,0.0,0.3,0.1]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",shadow=True,radius=1.1,labeldistance=1,startangle=90,textprops={"fontsize":13},
       counterclock=False,wedgeprops={"linewidth":3},rotatelabels=True)
# explode= the explode parameter in pie plot allows you to seperate and emphasize a specific pie by slice by pulling it away from the plot
# autopct=this will show the percentage of the table or list
# shadow=True ,and this parameter will gave the show of plot
# radius= value..., we can rise the size of plot using radius parameter
# lablesdistance = allows us to adjust distance between the slice and their corresponding lables  
# startangle = we can rotate it's angle in digree's like 60 30 90 180 degree's
# textprops = this parameter in plot allows you to customize the appearance of the text visualization
# counterclock = this parameter allows us to determine the direction in which the pie slices are drawn.defaultly it takes True but if we want to change 
#     then we can gave the false 
# wedgeprops= you can use the parameter to chaange the property of plot edge color,edge width or even add a shadow effect to the wedges
# rotatelabels=we can rotate the plots labels.it's work on true or false conditions 
plt.title("Pie Charts")
plt.legend(loc=4)
plt.show()
```

Dot pie plot

```python
x=[40,50,60,70,80]
x1=[80,70,60,50,40]
y=["C","C++","Java","Python","Javascript"]
ex=[0.0,0.0,0.0,0.3,0.1]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",shadow=True,radius=1.1,labeldistance=1,startangle=90,textprops={"fontsize":13},
       counterclock=False,wedgeprops={"linewidth":3},rotatelabels=True)
plt.pie([1],colors="w")
plt.title("Pie Plots")
plt.legend(loc=4)
plt.show()
```

Dot pie plot Second way

```python
x=[40,50,60,70,80]
y=["C","C++","Java","Python","Javascript"]
ex=[0.0,0.0,0.0,0.3,0.1]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",shadow=True,radius=1.1,labeldistance=1,startangle=90,textprops={"fontsize":13},
       counterclock=False,wedgeprops={"linewidth":3},rotatelabels=True)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)
plt.show()
```
